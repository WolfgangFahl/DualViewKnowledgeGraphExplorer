# Interactive SPARQL Query Runner

## Overview
This notebook provides an interactive interface to run SPARQL queries defined in the SEEKCommons resource-hub query_objects.yaml file.

## Setup
Install dependencies: `pip install pylodstorage pyyaml requests ipywidgets pandas`

## Configuration Files

**Query Definitions**
- Source: https://github.com/SEEKCommons/resource-hub/blob/main/documentation/query_objects.yaml
- Contains: Predefined SPARQL queries for SEEKCommons data
- Format: YAML with query names as keys

**Endpoint Configuration**
The notebook uses endpoint configurations that specify endpoint URL, rate limiting (calls_per_minute), and HTTP method (POST/GET).

## Usage

**Select Components**
1. Endpoint Selector: Choose SPARQL endpoint (e.g., wikidata)
2. Query Selector: Choose predefined query from query_objects.yaml
3. Limit Input: Set maximum number of results (default: 10)
4. Run Query Button: Execute the selected query

**Parameters**
Queries support parameters using $variable syntax in the YAML file. Currently supported: $limit for maximum number of results. Parameters are automatically replaced before query execution.

**Results**
Results are displayed as pandas DataFrames for easy analysis and visualization.

## Example Queries
- WhatLinksToSEEKCommons: Find Wikidata items linking to SEEKCommons
- SEEKCommonersWithORCIDs: List SEEKCommons members with ORCID identifiers
- SEEKCommonsWorkhops: Find workshops related to SEEKCommons

## Technical Details
Class: QuerySetRunner | SPARQL Library: pylodstorage | Query Language: SPARQL 1.1 | Parameter Substitution: $variable → actual value | Rate Limiting: Configurable per endpoint
